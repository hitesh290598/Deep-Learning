```python
# classification template

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the datasets
dataset = pd.read_csv('./Datasets/Churn_Modelling.csv')
```

```python
dataset
```

```python
dataset = dataset.drop(columns='Surname')
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset['Geography'] = le.fit_transform(dataset['Geography'])
dataset['Gender'] = le.fit_transform(dataset['Gender'])
```

```python
dataset
```

```python
X = dataset.iloc[:,1:-1]
Y = dataset.iloc[:,-1]
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
import keras

from keras.models import Sequential #to initalize NN
from keras.layers import Dense #to create layers
```

```python
classifier =Sequential()

classifier.add(Dense(output_dim=6, init='uniform', activation='relu', input_dim=11)) #1st hidden layer

classifier.add(Dense(output_dim=6, init='uniform', activation='relu')) #2nd hidden layer

classifier.add(Dense(output_dim=1, init='uniform', activation='sigmoid')) #output layer
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) #Compile the ANN 
```

```python
classifier.fit(X_train,Y_train, batch_size=11, nb_epoch=100)
```

```python
# create your classifier here

# Predicting the test set result
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_pred=(y_pred>0.5)

# Making Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,y_pred)
```
